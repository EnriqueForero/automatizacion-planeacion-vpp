## в™»пёЏ Medir tiempo

```python
import time
from datetime import timedelta

def medir_tiempo(nombre_seccion=""):
    """
    Crea un medidor de tiempo simple y elegante.

    ParГЎmetros:
    nombre_seccion (str): Nombre identificativo de la secciГіn a medir (opcional)

    Retorna:
    function: FunciГіn que al llamarla imprime el tiempo transcurrido
    """
    tiempo_inicio = time.time()

    def imprimir_tiempo():
        tiempo_total = time.time() - tiempo_inicio
        # Convertimos a un formato mГЎs legible usando timedelta
        tiempo_formateado = str(timedelta(seconds=tiempo_total))

        # Si se proporcionГі un nombre de secciГіn, lo incluimos en el mensaje
        seccion = f" para {nombre_seccion}" if nombre_seccion else ""
        print(f"Tiempo transcurrido{seccion}: {tiempo_formateado}")

    return imprimir_tiempo

# Ejemplo de uso:
"""
# Para medir una secciГіn especГ­fica:
fin_seccion = medir_tiempo("importar librerГ­as")
import pandas as pd
import numpy as np
fin_seccion()

# Para medir todo un bloque de cГіdigo:
fin_total = medir_tiempo()
# ... tu cГіdigo aquГ­ ...
fin_total()

# Para medir mГєltiples secciones:
fin_seccion1 = medir_tiempo("procesamiento inicial")
# ... cГіdigo primera secciГіn ...
fin_seccion1()

fin_seccion2 = medir_tiempo("anГЎlisis de datos")
# ... cГіdigo segunda secciГіn ...
fin_seccion2()
"""
```
